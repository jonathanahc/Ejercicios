## **Datos Temporales: datetime**

Los datos temporales son aquellos que representan el tiempo en distintas formas, como fechas, horas o combinaciones de ambas. Estos datos permiten analizar eventos en función del tiempo, detectar tendencias, predecir comportamientos y realizar cálculos como diferencias de tiempo o agrupaciones por períodos.

Los formatos relacionados a los datos temporales pueden ser:


1.   Fecha (date): 2025-02-28
2.   Tiempo/Hora(time): 18:00:00
3. Fecha y Hora (datetime): 2025-02-28 18:00:00

El uso de los datos temporales dependerá del objetivo de dicha variable. Esto determinará tanto su aplicación como su manejo.

1.   Si es **la clave principal** para analizar cómo se comporta otra variable durante el tiempo, entonces, su aplicación cae dentro del tema de **series temporales**. Por lo general, en este problema, los datos temporales son únicos y se manejan como índices. Tambièn resulta ùtil para acceder a datos grandes.
2.   Si no es el eje principal del análisis y solo es una variable más para extraer información temporal, se alamacena como una columna como cualquier otro dato. La ventaja de hacer esto es que se pueden aplicar diversas operaciones aritméticas.

En esta sesión nos enfocaremos en el punto 2, el cual resulta ùtil para el anàlisis descriptivo de los datos. Para ello, nos enfocaremos en el mòdulo/biblioteca `datetime`.

## Date, Time y Datetime

Existen diversos módulos o bibliotecas para manejar datos temporales en Python. Aquí usaremos `datetime`. Sin embargo, te envito a que revises los otros módulos que pudieran existir.

En `datetime`los tipos de datos y mòdulos que podemos encontrar son:


* `date(año, mes, dia)` : Devuelve un objeto de tipo date que representa la fecha con el año, mes y dia indicados.
* `time(hora, minutos, segundos, microsegundos)` : Devuelve un objeto de tipo time que representa un tiempo la hora, minutos, segundos y microsegundos indicados.
* `datetime(año, mes, dia, hora, minutos, segundos, microsegundos)` : Devuelve un objeto de tipo datetime que representa una fecha y hora con el año, mes, dia, hora, minutos, segundos y microsegundos indicados.
* `timedelta(days,seconds,microseconds,milliseconds,minutes,hours,weeks)`: Devuelve un objeto que representa la duración o diferencia entre dos fechas/horas.
* `tzinfo`: Se utiliza para representar la zona horario como timezone.utc.
* `timezone`: Se emplea para manejar zonas horarias  como la hora universal coordinada (utc).

Para usar la biblioteca `datetime` importamos primero y definimos un alias llamado dt:

```python
# Biblioteca con alias
import datetime as dt
```
Este tipo de importación es útil si se va a usar varias funciones como `timedelta`o `date`.

En la literatura seguramente encontrarás otra forma de importar la bibliteca:
``` python
# Biblioteca sin alias
from datetime import datetime
```

Esta resulta ideal si solo necesitas `datetime` y no otros módulos como `timedelta`o `date`.

En estos ejercicios vamos a usar el primer estilo. No obstante, el segundo también es válido.





````python
#Importamos la biblioteca con un alias
import datetime as dt

#fecha
print('Fecha ',dt.date(2025, 2, 28))

#Hora
print('Hora: ', dt.time(18, 12, 25))

#Fecha y Hora
print('Fecha y hora: ',dt.datetime(2025, 2, 28, 18, 12, 25))
````

## Acceso a los componentes de fecha y tiempo.

De cada uno de los objetos que representan fechas, tiempo y fecha-tiempo, es posible extraer algunos de sus componentes que lo integran como el año, mes, día, hora, entre otros.


### *Componentes de date*

Si tienes la siguiente instrucción `fecha=dt.date`los componentes que puedes extraer son:



* fecha.year : Devuelve el año de la variable llamada `fecha`, puede ser del tipo date o datetime.
* fecha.month : Devuelve el mes de `fecha`, que puede ser del tipo date o datetime.
* fecha.day : Devuelve el día de la fecha `fecha`, que puede ser del tipo date o datetime.
* fecha.weekday() : Devuelve el día de la semana de `fecha`, que puede ser del tipo date o datetime.

* date.today() : Devuelve un objeto del tipo date la fecha del sistema en el momento en el que se ejecuta.
* datetime.now(): Devuelve un objeto del tipo datetime con la fecha y la hora del sistema en el momento exacto en el que se ejecuta.

````python
#Importamos la biblioteca
import datetime as dt

fecha =dt.date(2025, 2, 28)

print('Año: ',fecha.year)
print('Mes: ',fecha.month)
print('Día: ',fecha.day)
print('Día de la semana: ',fecha.weekday())
print('Dia de la semana ISO: ',fecha.isoweekday())
````

### Otras Funciones de date

* `date.today()` : Devuelve un objeto del tipo date que contiene la fecha del sistema en el momento en el que se ejecuta.
* `date.fromisoformat("YYYY-MM-DD")`: Convierte una cadena en fecha
* `fecha.replace(year=X,month=Y,day=Z)`: reemplaza año, mes y dìa de la variable `fecha`
* `fecha.toordinal()`: Calcula el número de dìas desde el 1 de enero del año 1



````python
#imprime la fecha actual usando el alias dt y llamando a la clase date

hoy = dt.date.today()
print("Fecha actual:", hoy)

# Convertir cadena a objeto date
fecha = dt.date.fromisoformat("2025-03-01")
print("Fecha convertida:", fecha)  # Salida: 2025-03-01

nueva_fecha = fecha.replace(year=2028)
print("Fecha con nuevo año:", nueva_fecha)  # Salida: 2028-03-01

ordinal = fecha.toordinal()
print("Día ordinal:", ordinal)  # Salida: Número de días desde 01-01-0001
````

### Componentes de Time

Los componentes o atributos que se pueden extraer de `time` son:


* t.hour : Devuelve las horas del tiempo t, que puede ser del tipo time o datetime.
* t.minute : Devuelve los minutos del tiempo t, que puede ser del tipo time o datetime.
* t.second : Devuelve los segundos del tiempo t, que puede ser del tipo time o datetime.
* t.microsecond : Devuelve los microsegundos del tiempo t, que puede ser del tipo time o datetime.

Veamos ejemplos

````python
import datetime as dt

# Crear una hora específica (14:30:15)
mi_tiempo = dt.time(14, 30, 15, 23)

print("Hora:", mi_tiempo)           # Salida: 14:30:15
print("Hora:", mi_tiempo.hour)      # Salida: 14
print("Minuto:", mi_tiempo.minute)  # Salida: 30
print("Segundo:", mi_tiempo.second) # Salida: 15
print("Microsegundo: ",mi_tiempo.microsecond)
#
print("Esto es una cadena ",mi_tiempo.isoformat())
````


### Componentes de datetime: date+time+.tzinfo

`datetime`tiene los mismos componentes/atributos de `date` y `time`. En la siguiente código puedes ver los atributos que son similares.

````python
import datetime as dt
# Crear un objeto datetime
fecha_hora = dt.datetime(2024, 3, 1, 14, 30, 45, 123456)

# Extraer cada componente
print("Año:", fecha_hora.year)          # Salida: 2024
print("Mes:", fecha_hora.month)         # Salida: 3
print("Día:", fecha_hora.day)           # Salida: 1
print("Hora:", fecha_hora.hour)         # Salida: 14
print("Minuto:", fecha_hora.minute)     # Salida: 30
print("Segundo:", fecha_hora.second)    # Salida: 45
print("Microsegundo:", fecha_hora.microsecond)  # Salida: 123456
print("Día de la semana:", fecha_hora.weekday())     # Salida: 4 (Viernes)
print("Día de la semana ISO:", fecha_hora.isoweekday())  # Salida: 5 (Viernes)
````

*.tzinfo*

El atributo tzinfo en datetime.datetime se usa para manejar información de zonas horarias. Indica si un objeto datetime es "naive" (sin zona horaria) o "aware" (con zona horaria).

`.tzino` genera los siguiente:

* Fecha sin zona horaria (naive):	None
* Fecha con UTC
* Fecha con zona horaria específica. Por ejemplo, Nombre de la zona horaria

````python
import datetime as dt

# Crear un datetime sin zona horaria
fecha_naive = dt.datetime(2024, 3, 1, 14, 30)

# Verificar la zona horaria
print("Fecha sin zona horaria:", fecha_naive)
print("tzinfo:", fecha_naive.tzinfo)  # Salida: None
````

Ahora crearemos una fecha en la que indicaremos que tiene una zona horaria



````python
import datetime as dt

# Crear un datetime con zona horaria UTC
fecha_utc = dt.datetime(2024, 3, 1, 14, 30, tzinfo=dt.timezone.utc)

# Verificar la zona horaria
print("Fecha con UTC:", fecha_utc)
print("tzinfo:", fecha_utc.tzinfo)  # Salida: UTC
````

Se puede usar `replace`para cambiar la zona horaria de una fecha.

````python
import datetime as dt

# Crear un datetime con zona horaria UTC
# Fecha sin zona horaria (naive)
fecha_naive = dt.datetime(2025, 3, 1, 14, 30)

print("Fecha original zona horaria:", fecha_naive.tzinfo)  # Salida: None

# Asignar UTC a la fecha
fecha_utc = fecha_naive.replace(tzinfo=dt.timezone.utc)

print(fecha_utc)  # Salida: 2024-03-01 14:30:00+00:00
print("Fecha original zona horaria:", fecha_utc.tzinfo)  # Salida: None
````

## Otras representaciones de fechas, tiempo

Existen otras representaciones que marcan un tiempo o fecha. Una de ellas es timestamp

Un timestamp es un número que representa un instante en el tiempo como la cantidad de segundos (o milisegundos) transcurridos desde la época Unix (1970-01-01 00:00:00 UTC), es decir, se desea saber la fecha final con la cual se obtuvo el nùmero de segundos: fecha final-fecha unix.

A partir de este número, por ejemplo 1326244364, es posible obtener la fecha final. Para ello usamos `dt.timestamp(número)`.


````python
import datetime as dt

timestamp = dt.date.fromtimestamp(1326244364)
print("Fecha =", timestamp)
````

Si necesitas calcular el timestamp de una fecha específica, usas timestamp()

````python
import datetime as dt
fecha = dt.datetime(2025, 2, 28, 18, 00)  # Año, mes, día, hora, minuto
timestamp = fecha.timestamp()
print(timestamp)
````

## Convertir cadena a un objeto fecha

Para ello se usan una serie códigos para indicar el formato correcto, al cual se debería de convertir la fecha que está almacenada de tipo cadena.

Para saber los distintos códigos de formato consulta la pàgina de Python [aquì](https://docs.python.org/3/library/datetime-html#strftime-and-strptime-behavior)

````python
import datetime as dt
cadena_fecha = "2025-02-27"
formato = "%Y-%m-%d"  # Formato esperado

fecha_objeto = dt.datetime.strptime(cadena_fecha, formato)
print(fecha_objeto)  # Salida: 2025-02-27 00:00:00

cadena_fecha = "27-Feb-2025"
formato = "%d-%b-%Y"  # %b = Nombre corto del mes (Ene, Feb, Mar, etc.)

fecha_objeto = dt.datetime.strptime(cadena_fecha, formato)
print(fecha_objeto)  # Salida: 2025-02-27 00:00:00
````

````python
from datetime import datetime, date

# using date()
t1 = date(year = 2018, month = 7, day = 12)
t2 = date(year = 2017, month = 12, day = 23)

t3 = t1 - t2

print("t3 =", t3)

# using datetime()
t4 = datetime(year = 2018, month = 7, day = 12, hour = 7, minute = 9, second = 33)
t5 = datetime(year = 2019, month = 6, day = 10, hour = 5, minute = 55, second = 13)
t6 = t4 - t5
````

## Operaciones Aritméticas

Solo aplican para objetos `date`y `datetime`. Las operaciones que se pueden realizar son suma y resta.

Cuando se desea sumar o restar intervalos se usa `timestamp`.

Veamos ejemplos

````python
import datetime as dt

fecha1 = dt.date(2025, 3, 10)
fecha2 = dt.date(2025, 2, 27)

diferencia = fecha1 - fecha2
print(diferencia)         # 11 days, 0:00:00
print(diferencia.days)    # 11 (solo días)
````

````python
import datetime as dt

hora1 = dt.datetime(2025, 3, 10, 18, 30)
hora2 = dt.datetime(2025, 3, 10, 14, 15)

diferencia = hora1 - hora2  # timedelta
print(diferencia)  # Salida: 4:15:00
print(diferencia.total_seconds())  # Salida: 15300.0 (segundos)
````

### Operaciones Aritméticas con Timedelta

`timedelta(days, weeks,hours, minutes, seconds, etc.)` : Devuelve un objeto del tipo `timedelta` que representa un intervalo de tiempo con los dias, segundos y microsegundos indicados.

Con `timedelta(intervalo)`, podemos sumar a una *fecha base* o *fecha hora base* un determinado intervalo de tiempo expresado en días, horas, semanas, segundos, etc.



````python
import datetime as dt

fecha_base = dt.date(2025, 2, 27)
nueva_fecha = fecha_base + dt.timedelta(days=10)
print(nueva_fecha)  # 2025-03-09

fecha_anterior = fecha_base - dt.timedelta(weeks=2)
print(fecha_anterior)  # 2025-02-13


dt_base = dt.datetime(2025, 2, 27, 14, 30)

# Sumar 3 horas y 30 minutos
nueva_fecha_hora = dt_base + dt.timedelta(hours=3, minutes=30)
print(nueva_fecha_hora)  # 2025-02-27 18:00:00
````

````python
from datetime import datetime, date, timedelta

#
fecha_base = date(2025, 2, 27)
nueva_fecha = fecha_base + timedelta(days=10)
print(nueva_fecha)  # 2025-03-09

fecha_anterior = fecha_base - timedelta(weeks=2)
print(fecha_anterior)  # 2025-02-13


dt_base = datetime(2025, 2, 27, 14, 30)

# Sumar 3 horas y 30 minutos
nueva_fecha_hora = dt_base + timedelta(hours=3, minutes=30)
print(nueva_fecha_hora)  # 2025-02-27 18:00:00
````

## Convertir a Datetime Después de la Lectura: to_datetime

Podemos realizar la conversión una vez que se ha cargado el archivo. Para ello, usamos la instrucción `pd.to_datetime(df[columnas])``



````python

#cargamos la biblioteca pandas con un alias
import pandas as pd

df = pd.read_csv('drive/MyDrive/Unidad02/temporal01.csv')

df['FechaInicio']=pd.to_datetime(df['FechaInicio'])
df['FechaFin']=pd.to_datetime(df['FechaFin'])

print(df.head())
print(df.dtypes)
````

## pd.to_datetime() con formato específico

Si las fechas están en un formato poco común, podemos indicar el formato exacto con format.

Ejemplo con formato DD-MM-YYYY.

format='%d-%m-%Y' le dice a Pandas cómo interpretar la fecha (%d = día, %m = mes, %Y = año).
Evita errores de conversión cuando el formato no es detectado automáticamente.

````python
import pandas as pd
df2 = pd.DataFrame({
    'id_venta': [1, 2, 3],
    'fecha_venta': ['01-03-2025', '02-03-2025', '03-03-2025']
})

# Convertir con formato específico
df2['fecha_venta'] = pd.to_datetime(df2['fecha_venta'], format='%d-%m-%Y')

print(df2)
````

## Extraer Atributos/Componentes de los datos temporales

Tal y como se vió en secciones anteriores, en tipos de datos `datetime` se pueden extraer componentes como año, mes, día, semana, entre otros.

Veamos un ejemplo

````python
print('Dia: ',df['FechaFin'].dt.day)
print('Mes: ',df['FechaFin'].dt.month)
print('Año: ',df['FechaFin'].dt.year)
print('Hora: ',df['FechaFin'].dt.hour)
print('Minuto: ',df['FechaFin'].dt.minute)
````

## Operaciones Aritméticas

También podemos realizar operaciones aritméticas como sumar y restar. Veamos el siguiente ejemplo, donde usaremos las columnas `FechaInicio`y `FechaFin`.

En este sentido, es importante comentar que la operación de resta se puede hacer de manera directa. En el caso de la suma NO se puede. Para ello, es necesario utilizar `timedelta`.


````python
#De manera directa
print(df['FechaFin']-df['FechaInicio'])

#Cálculo del tiempo transcurrido desde la primera venta
print(df['FechaFin']-df['FechaInicio'].min())
````

````python
df["Diferencia"] = df["FechaFin"] - df["FechaInicio"]
# Diferencia en días
df["Diferencia_días"] = df["Diferencia"].dt.days

# Diferencia en horas
df["Diferencia_horas"] = df["Diferencia"].dt.total_seconds() / 3600

# Diferencia en minutos
df["Diferencia_minutos"] = df["Diferencia"].dt.total_seconds() / 60

print(df)
````

##  Agregar o Restar Tiempos a pd.Timestamp

Podemos sumar o restar días, horas o minutos a una fecha usando `pd.Timestamp`



````python
df['FechaEntrega'] = df['FechaFin'] + pd.Timedelta(days=3, hours=5)
print(df[['FechaFin', 'FechaEntrega']])
````

## Filtrado de Fechas
