# Scatter Trace: Gráfico de Dispersión o de Líneas

En este ejercicio vamos a generar un gráfico pequeño y simple. Para ello, usaremos un **Scatter Trace** que es un tipo de trazo (`trace`) utilizado para representar datos en un gráfico de dispersión o de líneas. La palabra `trace` se refiere a una serie de datos individuales que se visualizan en un gráfico.

En particular, un **Scatter Trace** (`go.Scatter`) se usa para dibujar puntos, lìneas o combinacciones de ambos en un espacio de coordenadas (**X, Y, Z, n gráficos 2D o 3D)**

## *Características un Scatter Trace (`go.Scatter`)*

Cuando se crea un scatter trace, se pueden configurar varias propiedades clave:

- `x` e `y`: Definen las coordenadas de los puntos en el gráfico.
- `mode`: Determina si el gráfico mostrará puntos (markers), líneas (lines) o ambos.
- `marker`: Personaliza el estilo de los puntos (tamaño, color, borde, transparencia, etc.).
- `line`: Personaliza las líneas (grosor, color, estilo, etc.).
- `name`: Etiqueta el conjunto de datos en la leyenda del gráfico.

**Opción uno:** Al crear el gráfico algunos programadores optan por crear una instancia vacía del objeto `Figure` y asignar a una variable `fig = go.Figure()`. Posteriormente, se añaden los demás elementos como el tipo de figura, nombre, los datos, entre otras, usando las instrucciones `add_trace(go.Scatter(parámetros))` y `update_layout(nombres)`.


**Una segunda opción:** Al crear la instancia `Figure` se define el tipo de figura (`go.Scatter`), los datos, el nombre, etc. `go.Figure(data=go.Scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13], mode='lines',name='Tendencia'))`

## **Scatter Trace con líneas (lines)**

En este ejemplo vamos a crear un gráfico de líneas (Scatter), con los siguientes argumentos:


1. `x=list(range(1, 11))` -> Define los valores del eje X (números dl 1 al 10).
2. `y = [i**2 for i in x]` -> Define los valores del eje Y (el cuadrado de cada valor de `x`
3. `mode='lines'`-> Especifica que quremos conectar los puntos con líneas, sin mostrar los marcadores
4. `name='x^2'`-> Asigna un nombre al trazo, que apacerá en la leyenda del gráfico
5. 'update_layout' -> Especifica el título del gráfico que aparecerá en la parte superior, así como el nombre de los ejes X e Y

````python
#Ejemplo 1 creando primero una instancia vacia del objeto Figure

import plotly.graph_objects as go

# Datos
x = list(range(1, 11))
y = [i**2 for i in x]

# Crear gráfico vacío
fig = go.Figure()

# Añadir trazo de líneas
#Se crea un gráfico de líneas utilizando go.Scatter. El argumento mode='lines'
#especifica que el gráfico debe mostrar una línea conectando los puntos de los datos.
# x e y son los datos a graficar, y name='x^2' asigna un nombre a la serie de datos.
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='x^2'))

#También se puede generar el gráfico incluyendo directamente
#los valors de x e y, dentro de go.scatter como puedes verlo en la línea de código
#siguiente que está comentada:
#fig.add_trace(go.Scatter(x = list(range(1, 11)), y=[i**2 for i in x], mode='lines', name='x^2'))


#title: Asigna un título general al gráfico.
#xaxis_title y yaxis_title: Asignan títulos a los ejes x y y, respectivamente.
# Títulos y etiquetas
fig.update_layout(
    title='Gráfico de Líneas Versión 1: y = x^2',
    xaxis_title='X',
    yaxis_title='Y'
)

#Muestra la figura
fig.show()
````

````python
#Ejemplo 2 al crear el objeto Figure se al mismo tiempo el trazo

import plotly.graph_objects as go

# Datos
x = list(range(1, 11))
y = [i**2 for i in x]

# Crear gráfico
#Se crea un gráfico de líneas utilizando go.Scatter. El argumento mode='lines'
# especifica que el gráfico debe mostrar una línea conectando los puntos de los datos.
# x e y son los datos a graficar, y name='x^2' asigna un nombre a la serie de datos.

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='x^2'))



#title: Asigna un título general al gráfico.
#xaxis_title y yaxis_title: Asignan títulos a los ejes x y y, respectivamente.
# Títulos y etiquetas
fig.update_layout(
    title='Gráfico de Líneas Versión 2: y = x^2',
    xaxis_title='X',
    yaxis_title='Y'
)

fig.show()
````

##**Scatter Trace con dos Líneas**

Este ejemplo crea un gráfico de líneas con dos conjuntos de datos (traces) en una misma figura. Resulta útil para comparar dos resultados. Veamos el desglose paso a paso.



1.   Importar la biblioteca: `import plotly.graph_objects as go`
2. Crear las variables con los datos de las dos líneas
3.   Crear una instancia del objeto `Figure`: `fig = go.Figure()`
4.  Agregar la primera serie de datos: `fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Serie 1 (x^2)'))`
4. Agregar la segunda serie de datos: `fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Serie 2 (x^2 + 1)'))`
5.  Configurar el diseño del gráfico: `fig.update_layout(...)`
6. Mostrar el gráfico: `fig.show()`


**Nota:** Interactúa con el gráfico al seleccionar algún región del  para hacer zoom. Con doble click regresa al estado original. Asimismo, pruebas las herramientas del **Modebar**



````python
#1. Importar la biblioteca
import plotly.graph_objects as go

# 2. Crear las variables con los datos de las dos líneas
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 5, 10, 17, 26]

# 3. Crear una instancia del objeto
fig = go.Figure()

# 4. Añadir primera serie (x vs y1)
#samos go.Figure() para crear el gráfico y luego añadimos dos series con fig.add_trace().
#La primera serie es x vs. y1 y la segunda es x vs. y2. Ambos son gráficos de líneas (mode='lines').
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Serie 1 (x^2)'))

#5. Añadir segunda serie (x vs y2)
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Serie 2 (x^2 + 1)'))

# 6. Títulos y etiquetas
fig.update_layout(
    title='Gráfico de Líneas con Varias Series',
    xaxis_title='X',
    yaxis_title='Y'
)

# 7. Mostrar el gráfico
fig.show()
````

##**Gráfico de Dispersión con Personalización de Puntos**

En este ejemplo, vamos a crear un gráfico de dispersión y personalizar los puntos con diferentes colores y tamaños. Utilizamos `numpy` para generar datos aleatorios y plotly.graph_objects (go) para construir la visualización. Los pasos son los siguientes:



1.   Importación de Librerias de Plotly y Numpy
2.   Generación de datos aleatorios que representa los puntos x e y, y otros que se usarán para asignar tamaños y colores a los puntos
3. Creación del gráfico de dispersión
4. Personalización del diseño
5. Mostrar el gráfico

**Nota:** Interactúa con el gráfico al seleccionar algún región del  para hacer zoom. Con doble click regresa al estado original. Asimismo, pruebas las herramientas del **Modebar**



````python
# 1. Importación de librerías de Plotly y Numpy
import plotly.graph_objects as go
import numpy as np

#2. Generación de datos aleatorios
np.random.seed(42) #Semilla para la reproducibilidad (siempre genera los mismos valores aleatorios)
x = np.random.rand(50) #50 valores entre 0 y 1 para el eje X
y = np.random.rand(50) #50 valores entre 0 y 1 para el eje Y
sizes = np.random.randint(10, 50, size=50)  # Tamaños aleatorios para los puntos
colors = np.random.rand(50)  # Colores aleatorios para los puntos

# 3. Crear gráfico de dispersión
#Usamos go.Scatter con mode='markers' para crear un gráfico de dispersión.
#Personalizamos los puntos con el parámetro marker, donde podemos cambiar su tamaño (size), color (color) y la paleta de colores (colorscale).
#El parámetro showscale=True muestra una barra de colores para representar los valores de color.

fig = go.Figure(data=go.Scatter(
    x=x, y=y, mode='markers',
    marker=dict(
        size=sizes,  # Tamaños de los puntos
        color=colors,  # Colores de los puntos
        colorscale='Viridis',  # Paleta de colores
        showscale=True  # Muestra la barra de colores
    )
))

# 4. Personalización del diseño: Títulos y etiquetas
fig.update_layout(
    title='Gráfico de Dispersión con Personalización de Puntos',
    xaxis_title='Eje X',
    yaxis_title='Eje Y'
)

#5. Mostrar el gráfico
fig.show()
````

##**Scatter Trace para Gráfico de Áreas Apiladas**

Un gráfico de áreas apiladas es útil para mostrar la evolución de varias series a lo largo del tiempo. Cada área está apilada sobre la anterior, mostrando la contribución acumulada.

````python
import plotly.graph_objects as go

# Datos
x = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
y1 = [10, 15, 20, 25, 30, 35]
y2 = [5, 10, 15, 20, 25, 30]
y3 = [2, 5, 8, 10, 12, 15]

# Crear gráfico de áreas apiladas
fig = go.Figure()

#Usamos go.Scatter con fill='tonexty' para crear un gráfico de áreas apiladas.
#El parámetro mode='none' se usa para evitar líneas de conexión entre los puntos.
#Las áreas se apilan encima de la anterior automáticamente.

# Añadir las tres series como áreas apiladas
fig.add_trace(go.Scatter(x=x, y=y1, fill='tonexty', mode='none', name='Serie 1'))
fig.add_trace(go.Scatter(x=x, y=y2, fill='tonexty', mode='none', name='Serie 2'))
fig.add_trace(go.Scatter(x=x, y=y3, fill='tonexty', mode='none', name='Serie 3'))

# Títulos y etiquetas
fig.update_layout(
    title='Gráfico de Áreas Apiladas',
    xaxis_title='Meses',
    yaxis_title='Valor'
)

fig.show()
````

# **Gráfico de Barras**

## **Gráficos de Barras Sencillo (Sintaxis Básica)**

Vamos a generar un gráfico de barras sencillo para mostrar la sintaxis básica de `go.Bar`. El siguiente código genera un gráfico en donde los datos representan diferentes frutas y sus cantidades.

````python
#1. Importar la biblioteca
import plotly.graph_objects as go

#2. Definir los datos
#Lista con los nombres d las frutas (valores del eje X)
categorias = ['Manzana', 'Plátano', 'Cereza', 'Naranja', 'Fresa']
#Lista con la cantidad de cada fruta (valores del eje Y)
cantidades = [10, 15, 7, 13, 9]

#3. Crear gráfico
#Se utiliza go.Bar para crear un gráfico de barras.
#Los datos x=categorias y y=cantidades se usan para representar las frutas y sus cantidades.
fig = go.Figure(data=go.Bar(x=categorias, y=cantidades))

#4. Personaliza el gráfico añadiendo títulos y etiquetas
fig.update_layout(
    title='Cantidad de Frutas',
    xaxis_title='Frutas',
    yaxis_title='Cantidad'
)

#5. Muestra el gráfico
fig.show()
````

##**Gráfico de Barras Apiladas Horizontal**

El gráfico de barras apiladas horizontal es una excelente opción para comparar categorías de manera visual y clara. En este tipo de gráficos se suman los valores en cada categoría.

En el siguiente ejemplo tenemos 3 grupos de resultados y 4 categorías. Para cada categoría se sumarán los resultados de cada grupo.

Si quieres que las barres no se apilen, sino que se coloquen una al lado de la otra, en lugar de sumarlas, cambia el valor de `barmode`de *'stack'* a *'group*.

````python
import plotly.graph_objects as go

# Datos
categorias = ['A', 'B', 'C', 'D']
valores1 = [20, 35, 50, 40]
valores2 = [10, 25, 15, 30]
valores3 = [5, 10, 20, 25]

# Crear gráfico de barras apiladas horizontal
fig = go.Figure()

# Añadir las series de barras
#Usamos go.Bar con orientation='h' para crear barras horizontales.
#La opción barmode='stack' en fig.update_layout() apila las barras.
#Esto permite comparar cómo se distribuyen los valores entre las diferentes categorías.

fig.add_trace(go.Bar(
    y=categorias, x=valores1, name='Valor 1', orientation='h'
))
fig.add_trace(go.Bar(
    y=categorias, x=valores2, name='Valor 2', orientation='h'
))
fig.add_trace(go.Bar(
    y=categorias, x=valores3, name='Valor 3', orientation='h'
))

# Títulos y etiquetas
fig.update_layout(
    title='Gráfico de Barras Apiladas Horizontal',
    xaxis_title='Valor',
    yaxis_title='Categorías',
    barmode='stack'  # Apila las barras
)

fig.show()
````

#**Gráfico 3D de Dispersión**

Un gráfico de dispersión 3D en Plotly se utiliza para visualizar datos en un espacio tridimensional (X, Y, Z). Es útil para representar relaciones entre tres variables y explorar patrones en los datos.

Se usa `go.Scatter3d()` para generar este tipo de visualización. Se deben proporcionar tres conjuntos de datos correspondientes a los ejes X, Y y Z.

En el siguiente ejercicio, además de la biblioteca de Plotly, se importa Numpy

1. Crea tres listas de datos, x, y, y z, representando coordenadas en el espacio 3D.
2. Usa go.Scatter3d para crear el gráfico de dispersión 3D.
3. Personaliza el gráfico con título y etiquetas.

Si usas el scroll del ratón sobre la figura generada podrás hacer zoom sobre la figura. De igual manera, si dejas presionado el click (izquierdo/derecho según sea tu configuración) podrás rotar la figura.

````python
import plotly.graph_objects as go
import numpy as np

# Datos
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Crear gráfico 3D
#go.Scatter3d crea un gráfico de dispersión en 3D.
#Los datos x, y y z son las coordenadas en 3D, y mode='markers' indica que los puntos deben representarse como marcadores.
fig = go.Figure(data=go.Scatter3d(x=x, y=y, z=z, mode='markers'))


# Títulos y etiquetas
fig.update_layout(
    title='Gráfico 3D de Dispersión',
    scene=dict(
        xaxis_title='Eje X',
        yaxis_title='Eje Y',
        zaxis_title='Eje Z'
    )
)

fig.show()
````

#**Mapa de Calor (Heatmap)**

Un mapa de calor (Heatmap) en Plotly es una representación visual de datos en dos dimensiones, donde los valores se muestran con una escala de colores. Es útil para analizar matrices de correlación, patrones en datos numéricos y tendencias en grandes volúmenes de información.

Se usa `go.Heatmap()` para generar un gráfico de mapa de calor. Este tipo de gráfico requiere tres elementos clave:

1. Valores en el eje X (Categorías o índices).
2. Valores en el eje Y (Categorías o índices).
3. Matriz de valores (Z), que representa los datos numéricos asignados a cada combinación de X e Y.

Ejecuta el código básico que se proporciona a continuación. En est ecódigo `data`continene los valors de X e Y.

*Etiquetas Personalizadas*
Podemos definir valores específicos en X y Y que aparecen en los ejes. Agrega al código básico lo siguiente



```python
x_labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
y_labels = ['Fila1', 'Fila2', 'Fila3', 'Fila4', 'Fila5', 'Fila6', 'Fila7', 'Fila8', 'Fila9', 'Fila10']
```

y en go.Figure, dentro de go.Heatmap:


```python
x=x_labels,  # Etiquetas en el eje X
y=y_labels,  # Etiquetas en el eje Y
```





````python
import plotly.graph_objects as go
import numpy as np

# Datos aleatorios para el mapa de calor
#data es una matriz de 10 x 10 de valores aleatorios
data = np.random.rand(10, 10)

# Crear mapa de calor
#Usamos go.Heatmap para crear un mapa de calor.
#z=data contiene los valores de la matriz, y colorscale='YlGnBu' define la paleta de colores.
#Puedes ver cómo el color de cada celda cambia según el valor en la matriz.

fig = go.Figure(data=go.Heatmap(
    z=data,  # Matriz de valores
    colorscale='YlGnBu'  # Escala de colores
))

# Títulos
fig.update_layout(
    title='Mapa de Calor',
    xaxis_title='Eje X',
    yaxis_title='Eje Y'
)

fig.show()
````

##**Visualizar una matriz de correlación con un mapa de calor**

En el siguiente ejemplo usamos Pandas para cagar un conjunto de datos, la conocida como iris. Calcularemos la correlación entre variables (todas numéricas) y se visualizará como un Heatmap.

Las correlaciones fuertes (positivas o negativas) se resaltarán con colores intensos



````python
import pandas as pd
import seaborn as sns

# Cargar un dataset de ejemplo
df = sns.load_dataset("iris")  # Conjunto de datos de flores

# Calcular la matriz de correlación solo para columnas numéricas
# Se excluye la columna 'species' que contiene strings
correlation_matrix = df.drop(columns=['species']).corr()

# Crear el mapa de calor
fig = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.index,
    colorscale='RdBu',  # Escala de colores para correlaciones positivas y negativas
    colorbar=dict(title="Correlación")
))

fig.update_layout(title="Matriz de Correlación")
fig.show()
````
