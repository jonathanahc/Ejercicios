# **Pandas**

¿Qué es Pandas?
* Biblioteca para análisis y manipulación de datos.
* Una gran importancia en el ecosistema de Python para la ciencia de datos.
*Principales estructuras de datos:
 - Series: unidimensional.
 - DataFrame: estructura de datos bidimensional. Se implementa en Python como una clase y cuando se crea una o hace referencia se le denomina objeto o instancia
* Proporcionan una forma estructurada y legible de manejar datos tabulares (filas y columnas).
* Soportan datos heterogéneos y operaciones específicas para análisis de datos.
* Ofrecen índices para una manipulación más intuitiva de datos.
* Fácil integración con otras bibliotecas como Scikit-learn y Matplotlib.
* Obtener datos por medeio de un API y crear un dataframe
* Exploración inicial y estadística descriptiva de conjuntos de datos.

### **Crear a partir de diccionarios, tuplas, listas**

Existen diversas formas de crear un dataframe.

````python
import pandas as pd

# Datos de estudiantes
data = {
    'Estudiante': ['Juan', 'María', 'Pedro', 'Ana'],
    'Matemáticas': [8, 9, 6, 7],
    'Lengua': [7, 9, 8, 6]
}

# Estudiante, Matemáticas y Lengua son los nombres de las columnas
# Agrega automáticamente  un indice numérico de forma secuencial
# Se le puede asignar un nombre a la columna de índice
# Crear DataFrame
df = pd.DataFrame(data)
print(df)

````

````python
import numpy as np
import datetime as dt

np.random.seed(0) # set seed so result is reproducible
datos =pd.DataFrame(
    {
        'random': np.random.rand(5),
        'text': ['hot', 'warm', 'cool', 'cold', None],
        'truth': [np.random.choice([True, False]) for _ in range(5)]
    },
    index=pd.date_range(
        end=dt.date(2019, 4, 21),
        freq='1D',
        periods=5,
        name='date'
    )
)

print(datos)
````

````python
#Se puede crear a partir de una lista de diccionario

datos = pd.DataFrame([
    {'mag': 5.2, 'place': 'California'},
    {'mag': 1.2, 'place': 'Alaska'},
    {'mag': 0.2, 'place': 'California'},
])

print(datos)
````

````python
#Se puede crear a partir de una lista de tuplas
list_of_tuples = [(n, n**2, n**3) for n in range(5)]
print(list_of_tuples)

datos =pd.DataFrame(
    list_of_tuples,
    columns=['n', 'n_squared', 'n_cubed']
)

print(datos)
````

## **Crear un dataframe usando un API**

Es una estrategia que se suele usar para extraer información de páginas web u otros recursos en línea. SIn embargo, no garantiza que podamos acceder, por restricciones del sitio, y obtener los mismo datos.

````python
import datetime as dt
import pandas as pd
import requests

yesterday = dt.date.today() - dt.timedelta(days=1)
api = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
payload = {
    'format': 'geojson',
    'starttime': yesterday - dt.timedelta(days=30),
    'endtime': yesterday
}
response = requests.get(api, params=payload)

# let's make sure the request was OK
#200 is an indicator was OK
response.status_code
````

Aislamos los datos de la respuesta JSON
Necesitamos verificar las estructuras de los datos de respuesta para saber dónde están nuestros datos.

````python
earthquake_json = response.json()
earthquake_json.keys()
````

La API del USGS proporciona información sobre nuestra solicitud en la clave de metadatos. Tenga en cuenta que el resultado será diferente, independientemente del rango de fechas que elija, porque la API incluye una marca de tiempo que indica cuándo se extrajeron los datos:

````python
earthquake_json['metadata']
````

Cada elemento de la matriz JSON es una fila de datos para nuestro dataframe.

````python
type(earthquake_json['features'])
````

Sus datos serán diferentes dependiendo de la fecha en que ejecute esto.

````python
earthquake_json['features'][0]
````

**Convertir a DataFrame**
Necesitamos extraer la sección de propiedades de cada entrada en la matriz JSON para crear nuestro dataframe.

````python
earthquake_properties_data = [
    quake['properties'] for quake in earthquake_json['features']
]
df = pd.DataFrame(earthquake_properties_data)
df.head()
````

**Agrupar datos**

Ejercicio: Si tuviéramos datos de diferentes cursos, puedes agrupar por curso y obtener la nota promedio en cada asignatura.

````python
# Crear DataFrame con curso
data_con_curso = {
    'Estudiante': ['Juan', 'María', 'Pedro', 'Ana', 'Luis'],
    'Curso': ['A', 'A', 'B', 'B', 'A'],
    'Matemáticas': [8, 9, 6, 7, 8],
    'Lengua': [7, 9, 8, 6, 8]
}
df_curso = pd.DataFrame(data_con_curso)

print(df_curso)

# Agrupar por curso y calcular el promedio por asignatura solo para las columnas numéricas
promedios_curso = df_curso.groupby('Curso').agg({
    'Matemáticas': 'mean',
    'Lengua': 'mean'
})

# Mostrar los promedios por curso
print(promedios_curso)
````

**Operaciones de conjunto**

Ejercicio: Usa el método merge para combinar dos DataFrames. Supón que tienes datos de estudiantes y sus cursos, y otro DataFrame con las notas finales.

````python
# Datos de estudiantes y cursos
data_estudiantes = {
    'Estudiante': ['Juan', 'María', 'Pedro', 'Ana'],
    'Curso': ['A', 'A', 'B', 'B']
}
df_estudiantes = pd.DataFrame(data_estudiantes)

# Datos de notas finales
data_notas = {
    'Estudiante': ['Juan', 'María', 'Pedro', 'Ana'],
    'Nota_Final': [7, 9, 6, 8]
}
df_notas = pd.DataFrame(data_notas)

# Unir los dos DataFrames por la columna 'Estudiante'
df_combinado = pd.merge(df_estudiantes, df_notas, on='Estudiante')
print(df_combinado)
````

**Calcular el promedio de cada estudiante.**

Ejercicio: Añadir una nueva columna Promedio que sea el promedio de las tres asignaturas para cada estudiante, muestra el promedio a dos digitos.

````python
print(df)
# Calcular el promedio por estudiante
df['Promedio'] = df[['Matemáticas', 'Lengua', 'Ciencias']].mean(axis=1)

# Redondear la columna "Promedio" a 2 decimales
df['Promedio'] = df['Promedio'].round(2)

# Mostrar el DataFrame con el nuevo promedio
print(df[['Estudiante', 'Promedio']])
````

**Encontrar estudiantes con alguna asignatura reprobada (nota menor a 5).**

Ejercicio: Filtra los estudiantes que tengan al menos una asignatura con una nota menor a 5.

````python
# Filtrar estudiantes con alguna asignatura reprobada
estudiantes_reprobados = df[(df['Matemáticas'] < 5) | (df['Lengua'] < 5) | (df['Ciencias'] < 5)]

# Mostrar los estudiantes reprobados
print(estudiantes_reprobados[['Estudiante', 'Matemáticas', 'Lengua', 'Ciencias']])
````

**Agrupar los estudiantes por curso y calcular el promedio por curso.**

Ejercicio: Agrupa a los estudiantes por su curso y calcula el promedio de las asignaturas para cada curso.

````python

# Asegurarnos de que las columnas de notas son numéricas (si hay algún error en el tipo de dato, convertimos a numérico)
df[['Matemáticas', 'Lengua', 'Ciencias']] = df[['Matemáticas', 'Lengua', 'Ciencias']].apply(pd.to_numeric, errors='coerce')

# Agrupar por 'Curso' y calcular el promedio por asignatura
promedios_curso = df.groupby('Curso')[['Matemáticas', 'Lengua', 'Ciencias']].mean()

# Mostrar los promedios por curso
print(promedios_curso)
````
