# **Subplots**

Los ejemplos de visualizaciones que hemos venido empleando muestran la forma de dibujar un sólo gráfico. En la clase sobre Seaborn (vea el video de Youtube) se mostró un ejemplo sobre crear varias figuras para poder realizar una comparativa. Para ello, se utilizaba la instrucción pairplot.

El mostrar dos gráficos o más, uno al lado del otro, permite compararlos visualmente. Esto da la ventaja de que el lector o espectador pueda entenderlo por si mismo.

Matplotlib permite crear varios gráficos en un mismo objeto (figura), organizándolos en filas y columnas. Esto recibe el nombre de subplots. Para ello, se emplean las funciones `subplot()` y `subplots()`.

La función para crear "subfiguras" en matplotlib se incluye una vez que se importa dicha biblioteca


```python
# matplotlib es la biblioteca
# pyplot es un submódulo
# plt es el alias
import matplotlib.pyplot as plt
```

La función `plt.subplot()` permite crear múltiples gráficos en una misma figura. Esto lo hace a través de una matriz/cuadrícula de gráficos, cuyo tamaño está determinado por el número de filas y columnas, donde cada celda contendrá una figura  que se puede personalizat de manera individual.

En la literatura existen dos formas diferentes de trabajar con subplots:

1. `plt.suplot(nrows, ncols,index)`: Enfoque imperativo para gráficos sencillos y rápidos, donde se usa plt de manera directa y la configuraciòn de cada subfigura está determinada por la función de llamada a cada figura.
2. `plt.suplots(nrows, ncols)`: Enfoque orientado a objetos que crea la cuadrícula vacía y devuelve un objeto `fig`y una matriz `ax` con referencias a cada subplot, por lo que, es posible modificar cada subfigura de manera independiente. Por ello, es útil para gráficos complejos.


## `plt.subplot(nrows,ncols,index)`


Esta función crea y activa un subplot individual en una cuadrícula de `nrows` × `ncols`, usando un índice (`index`), empieza en 1, que indica su posición.

En este enfoque, se debe de **llamar** `plt.subplot(nrows,ncols,index)`  **antes de cada gráfico**, en donde `index`indica la posición del subplot en la cuadrícula. Este forma resulta útil para gráficos rápidos y sencillos. Sin embargo, no existe una referencia directa a cada celda $[i,j]$, por lo que se usa directamente `plt`.



### Sintaxis Básica

Si queremos crear una figura que contenga dos gráficos uno al lado de el otro, es decir, una figura con 1 fila y 2 columnas, la sintaxis sería:

Para el primer gráfico
```python
# La figura tiene 1 fila y 2 columnas, y el primer gráfico
# estará en la posición 1
plt.subplot(1,2,1)

#Aquí debe ir la instrucción plt para crear el gráfico (barras, histogramas, líneas)
```
Para para el segundo gráfico
```python
# La figura tiene 1 fila y 2 columnas, y el segundo gráfico
# estará en la posición 2
plt.subplot(1,2,2)

#Aquí debe ir la instrucción plt para crear el gráfico (barras, histogramas, líneas)
```

Veamos un ejemplo:

````python
import matplotlib.pyplot as plt
import numpy as np

#Datos de la Figura 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#Figura 1
plt.subplot(1, 2, 1)
plt.plot(x,y)

#Datos de la Figura 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#Figura 2
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
````

Observa que cada vez que se usa `plt.subplot(nrows,ncols,index)` no cambia el número de filas y columnas, solo el índice.

Si queremos que las figuras aparezcan apiladas, es decir todas en una misma columna, sólo cambiamos el número de filas y columnas.

Para el primer gráfico
```python
# La figura tiene 2 filas y 1 columna, y el primer gráfico
# estará en la posición 1
plt.subplot(2,1,1)

#Aquí debe ir la instrucción usadon plt para crear el gráfico (barras, histogramas, líneas)
```
Para para el segundo gráfico
```python
# La figura tiene 2 filas y 1 columna, y el segundo gráfico
# estará en la posición 2
plt.subplot(2,1,2)

#Aquí debe ir la instrucción usadon plt para crear el gráfico (barras, histogramas, líneas)
```

````python
import matplotlib.pyplot as plt
import numpy as np

#Datos de la Figura 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])


#Figura 1
plt.subplot(2, 1, 1)
plt.plot(x,y)

#Datos de la Figura 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#Figura 2
plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
````

### Ajustar el tamaño de la cuadrícula y las subfiguras

Es posible definir el tamaño general de la cuadrícula, ya que como recordarás, se considera como una sola figura. Esto se puede hacer por medio de la instrucción `plt.figure()` cuyos argumentos son el número de unidads de ancho y de alto. Por ejemplo:  `plt.figure(figsize=(10, 5))` define el tamaño de la figura (cuadrícula) en 10 unidades de ancho y 5 de alto.

En el siguiente ejemplo creamos una figura que contiene 4 gráficos distribuidos en una cuadrícula de 2 x 2. En este código hemos añadido la instrucción `plt.figure(figsize=(10, 5))`. Ejecuta el código cambiando los valores de ancho y alto y observa lo que sucede

````python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 5))  # Tamaño de la figura


#Figura 1
#Datos
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#localización en la cuadrícula
plt.subplot(2, 2, 1)
plt.plot(x,y,linestyle = 'dotted')

#Figura 2
#Datos
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#localización en la cuadrícula
plt.subplot(2, 2, 2)
plt.plot(x,y,linestyle = 'dashed')

#Figura 3
#Datos
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#localización en la cuadrícula
plt.subplot(2, 2, 3)
plt.plot(x,y,'-o')

#Figura 4
#Datos
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

#localización en la cuadrícula
plt.subplot(2, 2, 4)
plt.scatter(x, y)


#plt.tight_layout()
plt.show()
````

Es posible que conforme cambies el tamaño de la caudrícula, aunque todas las subfiguras se ajustan, empezarás a observar que se superponen, principalmente en sus ejes. Para un ajuste automático de las todas ellas, se utiliza la instrucción `plt.tight_layout()`.

## `plt.subplots()`

Crea toda la cuadrícula (figura) de subplots de una sola vez y devuelve un objeto `fig` (figura) y una matriz `ax´ con referencias a cada subplot.

Usando `ax[row,col]`se puede acceder a cada subfigura individualmente, por lo que los gráficos se pueden modificar con más facilidad.

La sintaxis básica para crear una cuadrícula/matriz de 2 x 2, es la siguiente:


```python
# subplots()
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=2)
```

Esto crea una figura organizada en una cuadrícula de 2 x 2. Podemos acceder a cada posición de la figura como un arreglo por medio de `ax`. Es importante que observes que a diferencia del otro método de subfiguras (`subplot()`, en singular), los índices para cada elemento de la cuadrícula empiezan en cero (0).


```python
# 4 subfiguras
ax[0, 0].plot([1, 2, 3], [4, 5, 6])
ax[0, 1].scatter([1, 2, 3], [4, 5, 6])
ax[1, 0].bar([1, 2, 3], [4, 5, 6])
ax[1, 1].hist([1, 2, 3], bins=3)
```

Cada subfigura se puede personalizar de manera indepediente llamando los métodos correspondientes por medio de `ax`. Por ejemplo:


```python
# Títulos y etiquetas
ax[0, 0].set_title('Line Plot')
ax[0, 1].set_title('Scatter Plot')
ax[1, 0].set_title('Bar Chart')
ax[1, 1].set_title('Histogram
```

Veamos un ejemplo con histogramas.


````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 2, figsize=(10, 6))

data = np.random.randn(1000)

ax[0, 0].hist(data, bins=30, color="b", alpha=0.7)
ax[0, 0].set_title("Distribución Normal")

ax[0, 1].hist(np.random.uniform(-2, 2, 1000), bins=30, color="g", alpha=0.7)
ax[0, 1].set_title("Distribución Uniforme")

ax[1, 0].hist(np.random.exponential(1, 1000), bins=30, color="r", alpha=0.7)
ax[1, 0].set_title("Distribución Exponencial")

ax[1, 1].hist(np.random.poisson(3, 1000), bins=30, color="purple", alpha=0.7)
ax[1, 1].set_title("Distribución de Poisson")

plt.tight_layout()
plt.show()
````

### Compartir Ejes

Al crear múltiples gráficos en la misma figura en Matplotlib, es común querer compartir el eje x o y entre las subfiguras. Esto se puede hacer usando los parámetros `sharex` y `sharey` en la función `subplots()`.

En el siguiente ejemplo, se muestran dos subfiguras que comparten el eje X. Observa que solo uno de los gráficos muestran los valores de dicho eje, mientras que en la otra figura solo se observan los `ticks` sin leyenda alguna.

````python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plot on first subplot
ax1.plot([1, 2, 3], [4, 5, 6])
ax1.set_ylabel('Subplot 1')

# Plot on second subplot
ax2.plot([1, 2, 3], [7, 8, 9])
ax2.set_ylabel('Subplot 2')

# Set x-axis label for bottom subplot only
ax2.set_xlabel('X Label')

plt.show()
````

Podemos hacer lo mismo considerando ambos ejes. Ejecuta el siguiente código

````python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots(2, 2, figsize=(8, 6), sharex=True, sharey=True)

x = np.linspace(0, 10, 100)
ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.sin(x + np.pi/2), color="r")
ax[1,0].plot(x, np.sin(x))
ax[1,1].plot(x, np.sin(x + np.pi/2), color="g")

plt.show()
````

## GridSpec

`GridSpec` (del módulo `matplotlib.gridspec`) permite definir una cuadrícula flexible de subplots, donde podemos controlar el tamaño y la posición de cada gráfico en una figura.

`GridSpec` permite subplots de diferentes tamaños en una misma figura. Asimismo,  se puede combinar varias celdas para gráficos más grandes. Por lo que, ofrece más flexibilidad en la distribución de los subplots.

Para usar `GridSpec` , importamos la clase desde `matplotlib.gridspec`y la aplicamos a una figura.

En el siguiente ejemplo creamos una cuadrícula d 2 x 2, en donde, la tercera subfigura ocupa toda la segunda fila. Observa que para que esta figua ocupe todo este espacio se usa la instrucción `gs[1,:]`

````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(8, 6))  # Crear figura
gs = gridspec.GridSpec(2, 2)  # Definir una cuadrícula de 2x2

# Crear subplots en posiciones específicas
ax1 = fig.add_subplot(gs[0, 0])  # Primer subplot (fila 0, columna 0)
ax2 = fig.add_subplot(gs[0, 1])  # Segundo subplot (fila 0, columna 1)
ax3 = fig.add_subplot(gs[1, :])  # Tercer subplot ocupa toda la segunda fila

# Graficar en cada subplot
ax1.plot(x, np.sin(x))
ax1.set_title("Seno")

ax2.plot(x, np.cos(x))
ax2.set_title("Coseno")

ax3.plot(x, np.tan(x))
ax3.set_title("Tangente")
ax3.set_ylim(-10, 10)  # Limitar valores extremos

plt.tight_layout()
plt.show()
````

### Personalización de Tamaños

Podemos ajustar el ancho y alto relativo de los subplots con `width_ratios` y `height_ratios`.

Ejemplo con subfiguras de diferentes tamaños. Observa que usando `width_ratios=[2, 1, 1]`, El primer subplot es el doble de ancho que los demás, y con `height_ratios=[1, 2]`, la segunda fila es más alta.

````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0, 10, 100)
fig = plt.figure(figsize=(10, 6))
gs = gridspec.GridSpec(2, 3, width_ratios=[2, 1, 1], height_ratios=[1, 2])

# Crear subplots con diferentes tamaños
ax1 = fig.add_subplot(gs[0, 0])  # Más ancho
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
ax4 = fig.add_subplot(gs[1, :])  # Ocupa toda la segunda fila

# Graficar
ax1.plot(x, np.sin(x))
ax1.set_title("Seno")

ax2.plot(x, np.cos(x))
ax2.set_title("Coseno")

ax3.plot(x, np.tan(x))
ax3.set_title("Tangente")
ax3.set_ylim(-10, 10)

ax4.plot(x, np.exp(-x))
ax4.set_title("Exponencial")

plt.tight_layout()
plt.show()
````

### Fusionar Múltiples Celdas

Podemos hacer que una subfigura ocupe varias filas o columnas. En el siguiente ejemplo combinamos subfiguras en una figura 3x3, en donde:

* `gs[0, :2]`, la primera subfigura ocupa 2 columnas.
* `gs[1, :]`, la segunda subfigura ocupa toda la segunda fila.
* `gs[2, 1:]`,la última subfigura ocupa dos columnas.



````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(10, 6))
gs = gridspec.GridSpec(3, 3)

ax1 = fig.add_subplot(gs[0, :2])  # Ocupa dos columnas
ax2 = fig.add_subplot(gs[0, 2])
ax3 = fig.add_subplot(gs[1, :])  # Ocupa toda la fila 2
ax4 = fig.add_subplot(gs[2, 0])
ax5 = fig.add_subplot(gs[2, 1:])

# Graficar
ax1.plot(x, np.sin(x))
ax1.set_title("Seno - 2 columnas")

ax2.plot(x, np.cos(x))
ax2.set_title("Coseno")

ax3.plot(x, np.tan(x))
ax3.set_title("Tangente - fila completa")
ax3.set_ylim(-10, 10)

ax4.plot(x, np.exp(-x))
ax4.set_title("Exponencial")

ax5.plot(x, np.log(x))
ax5.set_title("Logaritmo")

plt.tight_layout()
plt.show()
````

## Agregando Espacios Entre Subplots (subplots_adjust)
Si los gráficos están demasiado juntos, podemos ajustar los espacios entre ellos.

````python
fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

fig.subplots_adjust(hspace=0.5, wspace=0.3)  # Espacio entre subplots

# Graficar
ax1.plot(x, np.sin(x))
ax1.set_title("Seno")

ax2.plot(x, np.cos(x))
ax2.set_title("Coseno")

ax3.plot(x, np.tan(x))
ax3.set_title("Tangente")
ax3.set_ylim(-10, 10)

ax4.plot(x, np.exp(-x))
ax4.set_title("Exponencial")

plt.show()
````

## Subplots en Seaborn

Debido a que Seabon es una librería basada en Matplotlib, también puede usar `plt.subplots()` para organizar múltiples gráficos.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Datos aleatorios
data = np.random.randn(1000)

# Crear la figura y los subplots (2 filas, 2 columnas)
fig, ax = plt.subplots(2, 2, figsize=(10, 6))

# Primer gráfico: Histograma normal
sns.histplot(data, bins=30, kde=True, ax=ax[0, 0])
ax[0, 0].set_title("Distribución Normal")

# Segundo gráfico: Uniforme
sns.histplot(np.random.uniform(-2, 2, 1000), bins=30, kde=True, ax=ax[0, 1])
ax[0, 1].set_title("Distribución Uniforme")

# Tercer gráfico: Exponencial
sns.histplot(np.random.exponential(1, 1000), bins=30, kde=True, ax=ax[1, 0])
ax[1, 0].set_title("Distribución Exponencial")

# Cuarto gráfico: Poisson
sns.histplot(np.random.poisson(3, 1000), bins=30, kde=True, ax=ax[1, 1])
ax[1, 1].set_title("Distribución de Poisson")

plt.tight_layout()
plt.show()
````

Seaborn tiene una función especial llamada `sns.FacetGrid()` que permite crear subplots automáticamente basados en categorías de datos.

````python
# Crear un DataFrame de ejemplo
df = sns.load_dataset("penguins")

# Crear subplots automáticos con `FacetGrid`
g = sns.FacetGrid(df, col="species", height=4)  # Un subplot por especie
g.map(sns.histplot, "flipper_length_mm")

plt.show()
````
